```python
def format_tick(x, pos):
    return f"{x:.3f}"

# Read the CSV file
data = pd.read_csv('./data/results_LU_unstable_8.csv')
df = pd.DataFrame(data, columns=['state_space_size', 'action_space_size', 'time2', 'time3', 'time4', 'error_norm', 'error_norm_LU'])
print(df)

# Pivot the dataframes
df_time2 = df.pivot('state_space_size', 'action_space_size', 'time2') / 1e9
df_time3 = df.pivot('state_space_size', 'action_space_size', 'time3') / 1e9
df_time4 = df.pivot('state_space_size', 'action_space_size', 'time4') / 1e9
df_speedup = df_time2 / df_time4
df_error_norm = df.pivot('state_space_size', 'action_space_size', 'error_norm')
df_error_norm_LU = df.pivot('state_space_size', 'action_space_size', 'error_norm_LU')

# Create a figure with subplots for each heatmap
fig, ((ax1, ax2, ax3, ax4)) = plt.subplots(1, 4, figsize=(25, 5))

# Subplot for rocket cycles 
sns.heatmap(df_time2, cmap="inferno", ax=ax1, norm=LogNorm(vmin=df_time2.min().min(), vmax=df_time2.max().max()), cbar_kws={'format': FuncFormatter(format_tick)})
ax1.set_xlabel('R')
ax1.set_ylabel('Q')
ax1.invert_yaxis()
ax1.set_title('Rocket Execution Time at 1GHz (s)')

# Subplot for gemmini cycles 
sns.heatmap(df_time4, cmap="inferno", ax=ax2, norm=LogNorm(vmin=df_time2.min().min(), vmax=df_time2.max().max()), cbar_kws={'format': FuncFormatter(format_tick)})
ax2.set_xlabel('R')
ax2.set_ylabel('Q')
ax2.invert_yaxis()
ax2.set_title('Rocket + Gemmini Execution Time at 1GHz (s)')

def format_tick(x, pos):
    return f"{x:.0f}"

# Subplot for avg_parallel_time
sns.heatmap(df_speedup, cmap="inferno", ax=ax3, norm=LogNorm(vmin=df_speedup.min().min(), vmax=df_speedup.max().max()), cbar_kws={'format': FuncFormatter(format_tick)})
ax3.set_xlabel('R')
ax3.set_ylabel('Q')
ax3.invert_yaxis()
ax3.set_title('HW Acceleration Speedup')

# Subplot for avg_parallel_time
sns.heatmap(df_error_norm_LU, cmap="inferno", ax=ax4, norm=LogNorm(vmin=df_error_norm_LU.min().min() + 1e-8, vmax=df_error_norm_LU.max().max()))
ax4.set_xlabel('R')
ax4.set_ylabel('Q')
ax4.invert_yaxis()
ax4.set_title('Result Error')

# Display the plots
plt.show()
plt.savefig("./data/plot_spike_heatmap.png")
```

```python
def format_tick(x, pos):
    return f"{x:.3f}"

# Read the CSV file
data = pd.read_csv('./data/results_condensed_8.csv')
df = pd.DataFrame(data, columns=['state_space_size', 'action_space_size', 'horizon_length', 'time2', 'time3', 'error_norm'])
print(df)

# Pivot the dataframes
df_time2 = df.pivot('state_space_size', 'action_space_size', 'time2') / 1e9
df_time3 = df.pivot('state_space_size', 'action_space_size', 'time3') / 1e9
df_speedup = df_time2 / df_time3
q = df.pivot('state_space_size', 'action_space_size', 'state_space_size')
r = df.pivot('state_space_size', 'action_space_size', 'action_space_size')
T = df.pivot('state_space_size', 'action_space_size', 'horizon_length')
flops = T*(4*q**3 + 6*q**2 + 2*q**2*r + 2*q**2 + r*q*r**2 + (2/3)*r**3+ r**2)
frac_perf = flops  / (df_time3 * 1e9 * 16)
df_time2 = df.pivot('state_space_size', 'action_space_size', 'time2') / 1e9
df_error_norm = df.pivot('state_space_size', 'action_space_size', 'error_norm')

# Create a figure with subplots for each heatmap
#fig, ((ax1, ax2, ax3, ax4)) =  plt.subplots(1, 4, figsize=(25, 5))
fig, ((ax1, ax2, ax3, ax4)) =  plt.subplots(1, 4, figsize=(25, 5))

# Subplot for rocket cycles 
sns.heatmap(df_time2, cmap="inferno", ax=ax1, norm=LogNorm(vmin=df_time2.min().min(), vmax=df_time2.max().max()), cbar_kws={'format': FuncFormatter(format_tick)})
ax1.set_xlabel('R Dimension')
ax1.set_ylabel('Q Dimension')
ax1.invert_yaxis()
ax1.set_title('Rocket Execution Time at 1GHz (s)')

# Subplot for gemmini cycles 
sns.heatmap(df_time3, cmap="inferno", ax=ax2, norm=LogNorm(vmin=df_time2.min().min(), vmax=df_time2.max().max()), cbar_kws={'format': FuncFormatter(format_tick)})
ax2.set_xlabel('R Dimension')
ax2.set_ylabel('Q Dimension')
ax2.invert_yaxis()
ax2.set_title('Rocket + Gemmini Execution Time at 1GHz (s)')

def format_tick(x, pos):
    return f"{x:.0f}"

# Subplot for avg_parallel_time
sns.heatmap(df_speedup, cmap="inferno", ax=ax3, norm=LogNorm(vmin=df_speedup.min().min(), vmax=df_speedup.max().max()), cbar_kws={'format': FuncFormatter(format_tick)})
ax3.set_xlabel('R Dimension')
ax3.set_ylabel('Q Dimension')
ax3.invert_yaxis()
ax3.set_title('HW Acceleration Speedup')

def format_tick(x, pos):
    return f"{x:.3f}"
# Subplot for avg_parallel_time
sns.heatmap(frac_perf, cmap="inferno", ax=ax4, norm=LogNorm(vmin=frac_perf.min().min(), vmax=frac_perf.max().max()), cbar_kws={'format': FuncFormatter(format_tick)})
ax4.set_xlabel('R Dimension')
ax4.set_ylabel('Q Dimension')
ax4.invert_yaxis()
ax4.set_title('Fraction of Peak Performance (Gemmini)')

# # Subplot for avg_parallel_time
# sns.heatmap(df_error_norm, cmap="inferno", ax=ax4, norm=LogNorm(vmin=df_error_norm.min().min() + 1e-8, vmax=df_error_norm_LU.max().max()))
# ax4.set_xlabel('R')
# ax4.set_ylabel('Q')
# ax4.invert_yaxis()
# ax4.set_title('Result Error')

# Display the plots
plt.savefig("./data/plot_spike_heatmap.pdf")
plt.show()
```
